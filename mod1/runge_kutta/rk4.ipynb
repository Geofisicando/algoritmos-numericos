# Método Runge-Kutta

Método numérico para a solução de equações diferenciais de primeira ordem.

Os métodos numéricos para equações diferenciais ordinárias são utilizados para encontrar aproximações para
soluções de EDOs. Eles também podem ser chamados de "integração numérica", mas neste sentido são mais gerais
do que o cálculo numérico de integrais.

As soluções da grande maioria das EDs não podem ser expressas em termos de funções elementares,
motivando o desenvolvimento e uso de métodos numéricos.

## Problema de valor inicial

```python
from IPython.display import Image
Image(filename="images/rk4_pvi.png",width=600)
```

Considere o seguinte problema de valor inicial. Estimar a função $f(x+h)$ com $f(x)$ conhecido.

$$\frac{dy}{dx}=f(x,y)$$

Integrando,

$$\int_{y_i}^{y_{i+1}} dy=\int_{x_i}^{x_{i+1}}f(x,y)dx$$

A integral à esquerda,

$$y_{i+1}-y_i=\int_{x_i}^{x_{i+1}}f(x,y)dx$$

Logo,

$$y_{i+1}=y_i+\int_{x_i}^{x_{i+1}}f(x,y)dx$$

Utilizando métodos numéricos:

$$y_{i+1}=y_i+h\phi$$

## O método Runge-Kutta de quarta ordem

Um membro da família de métodos Runge-Kutta é usado com tanta frequência que costuma receber o nome de "RK4" ou
simplesmente "Runge-Kutta". Para um problema de valor inicial:

$$\frac{dy}{dx}=f(x,y)$$

$$y(x_0)=y_0$$

Então o método RK4 para este problema será dado por:

$$y_{i+1}=y_i+h\frac{k_1+2k_2+2k_3+k_4}{6}$$

$$x_{i+1}=x_i+h$$

Então, o próximo valor ($y_{i+1}$) é dado pelo valor atual ($y_i$) somado com o tamanho do intervalo
($h$) e uma inclinação estimada ($\phi$). A inclinação é uma média ponderada das inclinações:

$$\phi=\frac{k_1+2k_2+2k_3+k_4}{6}$$

Onde:

$$k_1=f(x_0,y_0)$$

$$k_2=f(x_0+h/2,y_0+hk_1/2)$$

$$k_3=f(x_0+h/2,y_0+hk_2/2)$$

$$k_4=f(x_0+h,y_0+hk_3)$$

### Exemplo

Utilize o método Runge-Kutta para resolver $y'=y$ e $y(0)=1$ no intervalo $[0,1]$ com $h=0.1$.

Neste caso, $f(x,y)=y$. Então, para $n=0$, $x_0=0$ e $y_0=1$:

$$hk_1=hf(x_0,y_0)=hf(0,1)=(0.1)(1)=0.1$$

$$hk_2=hf(x_0+h/2,y_0+hk_1/2)=hf[0+\frac{1}{2}(0.1),1+\frac{1}{2}(0.1)]\\=hf(0.05,1.05)=(0.1)(1.05)=0.105$$

$$hk_3=hf(x_0+h/2,y_0+hk_2/2)=hf[0+\frac{1}{2}(0.1),1+\frac{1}{2}(0.105)]\\=hf(0.05,1.053)=(0.1)(1.053)=0.105$$

$$hk_4=hf(x_0+h,y_0+hk_3)=hf[0+(0.1),1+(0.105)]\\=hf(0.1,1.105)=(0.1)(1.053)=0.111$$

Logo,

$$y_1=y_0+\frac{1}{6}(hk_1+2hk_2+2hk_3+hk_4)=1+\frac{1}{6}[0.1+2(0.105)+2(0.105)+0.111]=1.105$$

e,

$$x_1=x_0+h=0+0.1=0.1$$

### Exercícios

1 - Utilize o método Runge-Kutta para resolver $y'=y-x$ e $y(0)=2$ no intervalo $[0,1]$ com $h=0.1$. Neste caso, $f(x,y)=y-x$ e $x_0=0$ e $y_0=2$.

Solução: $y(x)=e^x+x+1$

2 - Utilize o método Runge-Kutta para resolver $y'=y^2+1$ e $y(0)=0$ no intervalo $[0,1]$ com $h=0.1$. Neste caso, $f(x,y)=y^2+1$ e $x_0=0$ e $y_0=0$.

Solução: $y(x)=\tan x$
