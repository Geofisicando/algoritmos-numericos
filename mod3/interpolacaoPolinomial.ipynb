O nosso objetivo é encontrar o polinômio interpolador $P_1(x)$ que
passa pelos pontos $(x_0,y_0)$ e $(x_1,y_1)$:

$$f(x)\approx P_1(x)=a_0+a_1x$$

Com as seguintes condições,

$$P_1(x_0)=y_0$$

$$P_1(x_1)=y_1$$

Assim,

$$a_0+a_1x_0=y_0$$

$$a_0+a_1x_1=y_1$$

Colocando o sistema na forma matricial,

$$
\begin{bmatrix}
1 & x_0 \\
1 & x_1
\end{bmatrix}
\begin{bmatrix}
a_0 \\ a_1
\end{bmatrix}
=
\begin{bmatrix}
y_0\\ y_1
\end{bmatrix}
$$

Utilizando a Eliminação de Gauss:

$$
\begin{bmatrix}
1 & x_0 \\
0 & x_1-x_0
\end{bmatrix}
\begin{bmatrix}
a_0 \\ a_1
\end{bmatrix}
=
\begin{bmatrix}
y_0\\ y_1-y_0
\end{bmatrix}
$$

Daí, surge que,

$$a_1=\frac{y_1-y_0}{x_1-x_0}$$

Da primeira equação para $y_0$,

$$a_0+a_1x_0=y_0$$

Surge que,

$$a_0=y_0-x_0a_1$$

Substituindo na equação do Polinômio interpolador $P_1$,

$$P_1(x)=a_0+a_1x=(y_0-a_1x_0)+a_1x=y_0+a_1(x-x_0)$$

Por fim,

$$P_1(x)=y_0+\frac{y_1-y_0}{x_1-x_0}(x-x_0)$$

__Exemplo:__

Calcular $P_1(0.2)$ e $P_1(0.3)$ dados os valores,

$$x_0=0.1$$

$$y_0=1.221$$

$$x_1=0.6$$

$$y_1=3.32$$

Para estes valores, $P_1$ deverá ser:

$$P_1(0.2)=1.641$$

$$P_1(0.3)=2.061$$

```python
x0=0.1
y0=1.221
x1=0.6
y1=3.32

def polinomio_interp(x0,y0,x1,y1,x):
    return y0+((y1-y0)/(x1-x0))*(x-x0)

# Para x=0.2 P1(0.2)=1.641
p1=polinomio_interp(x0,y0,x1,y1,0.2)

print("P1(0.2)=%g"%p1)

# Para x=0.3 P1(0.2)=2.061
p1=polinomio_interp(x0,y0,x1,y1,0.3)

print("P1(0.3)=%g"%p1)
```
